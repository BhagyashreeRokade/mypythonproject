```
import random
#library that we use to choose random words from the list of words
 
name=input("What is your name? ")
#The user is asked to enter his/her name
 
print("Good luck!",name)
 
word=['apple','android','accenture','ball','bakery','bean','block',
'berry','blackberry','campus','candle','capital','cheese','ghost','host',
'computer','mobile','paper','clock','python','machine','security','threat',
'cyber','business','artificial','intelligence','system','support','science',
'binary','linear','model','structure']
 
 
word=random.choice(word)
#this will choose a random word from given list of words
    
print("Guess the characters")
guesses=''
    
attempts= 12
    
while attempts > 0:
    #counts the no. of times the user fails
    failed=0
        
    #for all characters from input
    #taking one at a time
    for char in word:
        #comparing the characters with characters in guesses
        if char in guesses:
            print(char)
                
        else:
            print('_')
            
            #for every failure 1 will be incremented in failure
            failed+=1
        
    if failed==0:
        #user will win if the failure is 0
        #And 'You Win!' will be the output
        print("You Win!")
            
        #this will print the correct word
        print("The word is: ",word)
        print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
        break
            
    #if user has input a wrong alphabet then 
    #it will ask user to enter another alphabet
    guess=input('Guess a character:')
       
    #every input character will be stored in guesses
    guesses+=guess
       
    #check input with the character in word
    if guess not in word:
           
        attempts-=1
           
        #if the charactr is not in the word 
        #"Wrong!" will be given as output
        print("Wrong!")
           
        #this will print the number of attempts remained for the user
           
        print('you have',+attempts,'more guesses')
           
        if attempts==0:
            print('You Loose!')
```
